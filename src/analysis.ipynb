```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../output/trainlog.csv")

#  3.1 Statystyki zbiorcze 
mean_ret        = df["Ret"].mean()
success_total   = df["Found"].sum()                # ile sukcesw cznie
success_ratio   = success_total / len(df)

# sukcesy na kade 10 000 krokw
df["Step_block"] = (df["Step"] // 10_000) * 10_000
succ_per_block   = (df.groupby("Step_block")["Found"]
                      .mean()   # odsetek
                      .mul(100))  # %
print(succ_per_block.head())

#  3.2 redni czas kroku w funkcji Episode 
avg_t = df.groupby("Episode")["t"].mean().reset_index()

fig_t = px.line(avg_t, x="Episode", y="t",
                title="redni czas kroku vs Episode",
                markers=True)
fig_t.show()
fig_t.write_image("reports/avg_t_vs_episode.pdf")   # statyczny PDF

#  3.3 Histogram Ret 
fig_ret = px.histogram(df, x="Ret", nbins=60,
                       title="Rozkad wartoci Ret")
fig_ret.show()
```
